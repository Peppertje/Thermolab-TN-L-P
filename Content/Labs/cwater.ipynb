## Veiligheid
We maken gebruik van een $10 \mathrm{\Omega}$, $10 \mathrm{W}$ weerstand.
Deze wordt snel heet.
De bronspanning mag dan ook alleen aan wanneer de weerstand in het water zit.
Raak de weerstand niet aan tijdens het experiment.
Omdat de weerstand in het water zit, kunnen we wel het elektrisch vermogen hoger zetten zonder dat de weerstand oververhit raakt.
Het maximaal vermogen mag $40 \mathrm{W}$ zijn.
Daarbij moet de roerder wel aanstaan om de warmte goed te verdelen.

## Data analyse

# Resultaten


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
t = np.linspace(0, 240, 17)
Q_1 = 14.4*1.49*t 
Q_2 = 40*t
data_T1 = np.array([17.7,18.2, 18.3, 18.6, 18.7, 19 ,19.1 ,19.3 ,19.5, 19.7 ,19.9 ,20.1 ,20.2 ,20.5, 20.6, 20.9 ,21])
data_T2 = np.array([17.3, 17.6, 18.1, 18.4, 18.8, 19.2, 19.5, 19.7, 20.2, 20.7, 20.9, 21.1, 21.6, 21.9, 22.4, 22.6, 23.1])
def Fit_func(T, c):
    return c*0.4*T*t
val, cov = curve_fit(Fit_func, data_T1, Q_1)
val1, cov1 = curve_fit(Fit_func, data_T2, Q_2)
print(val)
print(val1)
plt.figure()
plt.plot(t, Q_1)
plt.plot(t, Fit_func(data_T1, val))
plt.show()
````
